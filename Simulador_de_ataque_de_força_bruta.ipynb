```python
# =========================================
# 1. IMPORTAÇÕES
# =========================================
import time
import itertools
import string
import pandas as pd

# =========================================
# 2. USUÁRIO SIMULADO
# =========================================
usuario = "admin"

# =========================================
# 3. WORDLIST (ATAQUE DE DICIONÁRIO)
# =========================================
wordlist = [
    "123456",
    "password",
    "admin",
    "admin123",
    "qwerty",
    "abc123",
    "123",
    "abc",
    "senha123",
    "aB1!"
]

# =========================================
# 4. FUNÇÃO: ATAQUE DE DICIONÁRIO
# =========================================
def ataque_dicionario(usuario, senha_real, wordlist):
    tentativas = 0
    inicio = time.time()

    for senha in wordlist:
        tentativas += 1

        if senha == senha_real:
            fim = time.time()
            return {
                "metodo": "Dicionário",
                "senha_encontrada": senha,
                "tentativas": tentativas,
                "tempo": fim - inicio
            }

    return {
        "metodo": "Dicionário",
        "senha_encontrada": None,
        "tentativas": tentativas,
        "tempo": time.time() - inicio
    }

# =========================================
# 5. FUNÇÃO: BRUTE FORCE
# =========================================
def brute_force(senha_real, caracteres, max_len=4):
    tentativas = 0
    inicio = time.time()

    for tamanho in range(1, max_len + 1):
        for tentativa in itertools.product(caracteres, repeat=tamanho):
            tentativas += 1
            tentativa = ''.join(tentativa)

            if tentativa == senha_real:
                fim = time.time()
                return {
                    "metodo": "Brute Force",
                    "senha_encontrada": tentativa,
                    "tentativas": tentativas,
                    "tempo": fim - inicio
                }

    return {
        "metodo": "Brute Force",
        "senha_encontrada": None,
        "tentativas": tentativas,
        "tempo": time.time() - inicio
    }

# =========================================
# 6. FUNÇÃO: CLASSIFICAR COMPLEXIDADE
# =========================================
def classificar_senha(senha):
    if len(senha) <= 3:
        return "Fraca"
    elif len(senha) <= 5:
        return "Média"
    else:
        return "Forte"

# =========================================
# 7. SENHAS PARA TESTE
# =========================================
senhas_teste = [
    "123",
    "abc",
    "admin",
    "aB1!",
    "Zx9@",
    "senha123",
    "Aa1@",
]

# =========================================
# 8. EXECUÇÃO DAS SIMULAÇÕES
# =========================================
resultados = []
caracteres = string.ascii_letters + string.digits + "!@#"

for senha_real in senhas_teste:

    # Ataque de dicionário
    resultado_dic = ataque_dicionario(usuario, senha_real, wordlist)
    resultado_dic["senha_real"] = senha_real
    resultado_dic["complexidade"] = classificar_senha(senha_real)
    resultados.append(resultado_dic)

    # Ataque brute force
    resultado_brute = brute_force(senha_real, caracteres, max_len=4)
    resultado_brute["senha_real"] = senha_real
    resultado_brute["complexidade"] = classificar_senha(senha_real)
    resultados.append(resultado_brute)

# =========================================
# 9. CRIAR DATAFRAME
# =========================================
df = pd.DataFrame(resultados)

df = df[
    [
        "senha_real",
        "complexidade",
        "metodo",
        "senha_encontrada",
        "tentativas",
        "tempo"
    ]
]

print("Prévia dos dados:")
display(df)

# =========================================
# 10. EXPORTAR PARA EXCEL
# =========================================
arquivo = "resultado_bruteforce.xlsx"
df.to_excel(arquivo, index=False)

print(f"\nArquivo '{arquivo}' gerado com sucesso!")

# =========================================
# 11. DOWNLOAD DO ARQUIVO
# =========================================
from google.colab import files
files.download(arquivo)
```
